```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
cv.get_feature_names()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
new_df['tags'][0]
```

```python
stem('a young woman joins forces with a ufo enthusiast to investigate her boyfriend’s sudden disappearance and stumbles into a wild conspiracy. jeonyeo-been nana leedong-hwi rohdeok drama comedy mystery sci-fi&fantasy friendship investigation ufo miniseries suspense')
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity
```

```python
new_df[new_df['name']=='Glitch'].index[0]
```

```python
def recommend(kdrama):
    drama_index = new_df[new_df['name']==kdrama].index[0]
    distances = similarity[drama_index]
    drama_list = sorted(list(enumerate(distances)),reverse = True, key = lambda x : x[1])[1:6]
    
    for i in drama_list:
        print(new_df.iloc[i[0]]["name"])
```

```python
recommend('Goblin')
```
